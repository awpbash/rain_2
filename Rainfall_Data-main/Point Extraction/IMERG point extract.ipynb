```python
import os
import pandas as pd
import numpy as np
import xarray as xr
from pysteps.downscaling import rainfarm

# Define the folder containing NetCDF files
nc_folder_path = r"C:\Users\userAdmin\Desktop\Rain data files\IMERG 10yrs\data files"
seasons = {
    'DJF': [12, 1, 2],
    'MAM': [3, 4, 5],
    'JJA': [6, 7, 8],
    'SON': [9, 10, 11]
}

# Convert the selected stations to coordinates of interest
coordinates_of_interest = [(station[2], station[1]) for station in selected_stations]

print("Selected coordinates of interest:", coordinates_of_interest)

# Function to determine the season
def get_season(month):
    for season, months in seasons.items():
        if month in months:
            return season
    return None
# Initialize an empty DataFrame to store the results
columns = ['Datetime', 'Season'] + [station[0] for station in selected_stations]
results = pd.DataFrame(columns=columns)

n = 0
total = len(os.listdir(nc_folder_path))

# Iterate over all NetCDF files in the folder
for nc_file in os.listdir(nc_folder_path):
    n += 1
    print(f"{n}/{total}")
    if nc_file.endswith('0500.nc') and nc_file[:4] > "2015" and nc_file[:4] < "2017": #<--- Change the year range here
        print(f"Processing file {nc_file}...")
        try:
            # Construct the full path to the NetCDF file
            file_path = os.path.join(nc_folder_path, nc_file)
            
            # Open the NetCDF file
            ds = xr.open_dataset(file_path, engine='netcdf4')
            
            # Extract the time, latitude, and longitude data
            time_data = ds['time'].values
            lat_data = ds['lat'].values
            lon_data = ds['lon'].values
            
            # Prepare data storage for each coordinate of interest
            location_data = {i: [] for i in range(len(coordinates_of_interest))}
            
            # Iterate over each time step
            for t in time_data:
                date = pd.to_datetime(str(t))
                month = date.month
                season = get_season(month)
                
                # Extract data for each coordinate of interest
                row = {'Datetime': date, 'Season': season}
                for idx, (lon_of_interest, lat_of_interest) in enumerate(coordinates_of_interest):
                    # Find the nearest grid point to the specified latitude and longitude
                    lat_idx = np.abs(lat_data - lat_of_interest).argmin()
                    lon_idx = np.abs(lon_data - lon_of_interest).argmin()
                    
                    # Extract the rainfall data for the nearest grid point
                    rainfall_data = ds['precipitation'].sel(time=t, method='nearest').values
                    row[selected_stations[idx][0]] = rainfall_data[lat_idx, lon_idx]
                
                # Append the row to the results DataFrame
                results = pd.concat([results, pd.DataFrame([row], columns=columns)], ignore_index=True)
            
            ds.close()  # Close the dataset after processing
        except Exception as e:
            print(f"Error processing file {nc_file}: {e}")

# Save the results to a CSV file
results.to_csv('IMERG_Point_extracted.csv', index=False)

# Display the first few rows of the results DataFrame
print(results.head())
```
